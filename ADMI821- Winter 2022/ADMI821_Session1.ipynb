Read the Data:

```python
import requests

download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
response = requests.get(download_url)
open("nba_all_elo.csv", "wb").write(response.content) # "wb": write and binary

nba = pd.read_csv("nba_all_elo.csv") # other file types
# type(nba)
display(type(nba))
display(len(nba))
display(nba.shape) # dimensions of dataset (rows, cols)
nba.tail()
nba.head()
```

```python
pd.set_option("display.max.columns", None) # to see all the columns
pd.set_option("display.precision", 2) # show two decimals
nba.tail(3)
```
